```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
from sklearn.datasets import fetch_california_housing
from sklearn.manifold import TSNE

# Global Vars
SEED = 421067
# Warning: heavy computation
# 5-10 min
CALCULATE_TSNE = False
PERPLEXITY = 17
```

```python
def get_houses():
    X, y = fetch_california_housing(return_X_y=True)
    #y = np.log(y)
    X = np.concatenate([X, y[..., np.newaxis]], axis=1)
    return X
    
def plot_clusters_california(x, y, clusters,
                             title='California Houses Clusters: t-SNE',
                             alpha=1/3):
    fig, ax = plt.subplots(figsize=(18,14))
    #fig.suptitle('California Houses Clusters', fontsize=32)
    ax.set_title(title, fontsize=32)
    scatter = ax.scatter(x, y, c=clusters, alpha = 1/3)
    #ax.set_xticklabels([])
    #ax.set_yticklabels([])
    # Produce a legend with the unique colors from the scatter
    legend1 = ax.legend(*scatter.legend_elements(),
                       loc='lower left', title='Clusters', fontsize=20)
    ax.add_artist(legend1)
    title = legend1.get_title()
    title.set_fontsize(20)
```

```python
X = get_houses()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Warning: heavy computation
# 5-10 min

if CALCULATE_TSNE:
    visualization_axes = TSNE(n_jobs = -1,
                              perplexity=PERPLEXITY,
                              random_state=SEED).fit_transform(X)
    tsne_x = visualization_axes[:,0]
    tsne_y = visualization_axes[:,1]
    # Saving the calculation to disk
    file = open('transformation_x_y', 'wb')
    pickle.dump((tsne_x, tsne_y), file)
    file.close()
else:
    file = open('transformation_x_y','rb')  
    # load the object from the file into var b
    tsne_x, tsne_y = pickle.load(file)
    file.close()
```

```python
from sklearn.cluster import AgglomerativeClustering

a_agg = AgglomerativeClustering(memory='agg_tree',
                                compute_full_tree=True,
                                affinity='euclidean',
                                linkage='ward')
```

```python
clusters_agg4 = a_agg.set_params(n_clusters=6).fit_predict(X)
```

```python
plot_clusters_california(tsne_x, tsne_y, clusters_agg4)
```

```python
X_raw = get_houses()
```

```python
header=('MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'Latitude', 'Longitude', 'MedPrice')
houses = pd.DataFrame(X_raw, columns=header)
houses['Cluster'] = clusters_agg4
del X_raw
```

```python
houses.to_csv('cal_houses.csv')
```
